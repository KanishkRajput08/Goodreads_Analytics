## Overview of the Dataset on Best Books of the 20th Century
**This dataset comprises information about the best books published during the 20th century (from January 1st, 1901, to December 31st, 2000), scraped from Goodreads.com using Beautiful Soup. It includes 7597 books as rows and 16 columns. For each book, data such as the book's title, author's name, average rating, overall rating, publication date, description, total pages, genres, author's followers on Goodreads, total number of reviews, total number of books by the author, and star ratings are collected. This dataset offers a comprehensive view of the literary landscape during the 20th century, providing insights into popular books, authors, genres, reader preferences, and the impact of technological advancements on literature during that era.**

## *Summary of Cleaned Book Dataset:-*
**Contains 7597 Rows and 15 Columns**
- **title:** The name of the book.
- **authors:** Author name.
- **rating:** Total number of Rating for each book.
- **average_rating:** Number of Average Rating for each book.
- **date_of_publication:** Year in which the book was first published.
- **total_pages:** Number of pages in the book.
- **genres:** The genre the book belongs to (a style or category of art, music, or literature).
- **author_followers:** Number of followers of the author on the website.
- **no_of_reviews:** Total number of reviews given to the book.
- **no_of_books_by_author:** Number of books written by that author.
- **one_star_ratings:** Number of 1-star ratings out of 5 (Worse).
- **two_star_ratings:** Number of 2-star ratings out of 5 (Bad).
- **three_sta_ratings:** Number of 3-star ratings out of 5 (Average).
- **four_star_ratings:** Number of 4-star ratings out of 5 (Best).
- **five_star_ratings:** Number of 5-star ratings out of 5 (Excellent).

```python
# Dropping the description column as it currently contains 98 null values and is not crucial for the analysis at this stage
books.drop('descriptions', axis=1, inplace = True)
```

```python
# Dropping rows with null values from the Books dataset.
books.dropna(inplace = True)
```

```python
# Removing commas from the 'author_followers' column in the Books dataset.
books['author_followers'] = books['author_followers'].str.replace(',','')
```

```python
# Extracting the first word from the 'author_followers' column in the Books dataset
books['author_followers'] = books['author_followers'].str.split().str.get(0)
```

```python
# Converting 'author_followers' values ending with 'k' to actual numeric values in the Books dataset.
condition = books['author_followers'].str.endswith('k')
books.loc[condition, 'author_followers'] = (books.loc[condition, 'author_followers'].str.split('k').str.get(0).astype(float) * 1000)
```

```python
# Converting 'author_followers' column to integer data type in the Books dataset.
books['author_followers'] = books['author_followers'].astype(int)
```

```python
# Cleaning and converting the 'no_of_reviews' column to integer data type in the Books dataset.
books['no_of_reviews'] = books['no_of_reviews'].str.split().str.get(0).str.replace(',','')
```

```python
# Converting the 'no_of_reviews' column to integer data type in the Books dataset.
books['no_of_reviews'] = books['no_of_reviews'].astype(int)
```

```python
# Removing commas from the 'no_of_books_by_author' column in the Books dataset.
books['no_of_books_by_author'] = books['no_of_books_by_author'].str.replace(',','')
```

```python
# removing 'k' from the end and multiply by 1000 instead from the the no_of_books_by_author
condition = books['no_of_books_by_author'].str.endswith('k')
books.loc[condition, 'no_of_books_by_author'] = books[books['no_of_books_by_author'].str.contains('k')]['no_of_books_by_author'].str.replace('k','').astype(float)*1000
```

```python
# changing the datatype of column no_of_books_by_author from object to int
books['no_of_books_by_author'] = books['no_of_books_by_author'].astype('int64')
```

```python
# removing rating from the values
books['rating'] = books['rating'].str.split().str.get(0)
```

```python
# removing comma's from the values
books['rating'] = books['rating'].str.replace(',','')
```

```python
# changing the datatype of rating into int64
books['rating'] = books['rating'].astype('int64')
```

```python
# Cleaning and converting the 'one_star_ratings' column to integer data type in the Books dataset.
books['one_star_ratings'] = books['one_star_ratings'].str.split().str.get(0).str.replace(',','').astype(int)
```

```python
# Cleaning and converting the 'two_star_ratings' column to integer data type in the Books dataset.
books['two_star_ratings'] = books['two_star_ratings'].str.split().str.get(0).str.replace(',','').astype(int)
```

```python
# Cleaning and converting the 'three_star_ratings' column to integer data type in the Books dataset.
books['three_star_ratings'] = books['three_star_ratings'].str.split().str.get(0).str.replace(',','').astype(int)
```

```python
# Cleaning and converting the 'four_star_ratings' column to integer data type in the Books dataset.
books['four_star_ratings'] = books['four_star_ratings'].str.split().str.get(0).str.replace(',','').astype(int)
```

```python
# Cleaning and converting the 'five_star_ratings' column to integer data type in the Books dataset.
books['five_star_ratings'] = books['five_star_ratings'].str.split().str.get(0).str.replace(',','').astype(int)
```

```python
# Removing commas from the 'date_of_publication' column in the Books dataset.
books['date_of_publication'] = books['date_of_publication'].str.replace(',','')
```

```python
# Removing 'pagesPublished' text from the 'date_of_publication' column in the Books dataset.
condition = books['date_of_publication'].str.contains('pagesPublished')
books.loc[condition, 'date_of_publication'] = books[books['date_of_publication'].str.contains('pagesPublished')]['date_of_publication'].str.split('d ').str.get(-1).str.strip()
```
